```python
import regex as re
import numpy as np
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
import matplotlib
%matplotlib inline
from matplotlib import pyplot as plt
plt.tight_layout()
# plt.rcParams['figure.figsize'] = [12, 8]
# plt.rcParams['figure.dpi'] = 100
import seaborn as sns
sns.set(color_codes=True)
sns.set_style('white')
sns.set_context("paper", font_scale = 2)

ROOT = "/mnt/lerna/models/"
df = pd.read_csv("scripts/approximation-stats.csv")
df.fillna("", inplace=True)
df["method"] = df.apply(lambda row: row["approximator"] + str(row["beam width"]), axis=1)
df = df.sort_values(by=["model", "method"],ascending=True)
markers = ["o","*", ".", "D", "X", "v", "^"][:len(df["model"].unique())]
palette = sns.color_palette(None, len(df["method"].unique()), desat=1) 
df
```

## Plot SCE over calibration for each unique method

This ensures the intuition that SCE decreases if we use a post-hoc calibration method

```python
calibration_index = {v:k for k,v in dict(enumerate(df["calibrator"].unique().tolist())).items()}

metric = "MCE"
extended = 0.5
for m in sorted(df["method"].unique()):
    mdata = df[df["method"] == m]
    mdata["cal"] = mdata["calibrator"].apply(lambda x: calibration_index.get(x))
    
    #now can compare over models and calibrations
    sns.lmplot(data=mdata, x="cal", y=metric, hue="model", scatter=True, fit_reg=True, ci=None, legend=True,scatter_kws={"s": 50, "alpha":0.7},markers=markers,palette=palette)
    plt.title(m + "-" + metric)
    plt.xlim(-extended, max(calibration_index.values())+extended)
    plt.savefig(m + "-" + metric)
```

## Plot variance for each approximator 
[within a given model]

```python
model = "conll_03-BERT_NER_biose_all_dice"
calibration_index = {v:k for k,v in dict(enumerate(df["calibrator"].unique().tolist())).items()}

metric = "variance"
extended = 0.5
#for m in sorted(df["method"].unique()):
mdata = df[df["model"] == model]
mdata["cal"] = mdata["calibrator"].apply(lambda x: calibration_index.get(x))

#now can compare over models and calibrations
sns.lmplot(data=mdata, x="diversity", y=metric, hue="method", scatter=True, fit_reg=True, ci=None, legend=True,scatter_kws={"s": 50, "alpha":0.7})
           #,markers=markers,palette=palette)
    
plt.title("variance over estimators over diversity")
plt.xlim(-extended, 1+extended)
#     plt.savefig(m + "-" + metric)
```

```python
g = sns.catplot(data=mdata, x="diversity", y=metric, hue="method",legend_out=False,kind="box")
g.add_legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.,frameon=False)
g.ax.tick_params(axis='x', labelrotation=90)
plt.title("var over estimators")
```
